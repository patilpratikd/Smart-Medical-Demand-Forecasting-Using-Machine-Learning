```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Dell Latitude 5490/Downloads/Pharmacy.csv')
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
numerical_cols = ['Quantity', 'ReturnQuantity', 'Final_Cost', 'Final_Sales']

for col in numerical_cols:

    plt.hist(df[col], bins=50)

    plt.title(f'Histogram of {col}')

    plt.show()
```

```python
for col in numerical_cols:

    plt.boxplot(df[col])

    plt.title(f'Boxplot of {col}')

    plt.show()
```

```python
scatter_cols = [('Quantity', 'ReturnQuantity'), ('Final_Cost', 'Final_Sales')]

for col1, col2 in scatter_cols:

    plt.scatter(df[col1], df[col2])

    plt.xlabel(col1)

    plt.ylabel(col2)

    plt.title(f'Scatter Plot of {col1} vs {col2}')

    plt.show()
```

```python
corr_matrix = df.corr()

print(corr_matrix)
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix')

plt.show()
```

```python
df['Formulation'].fillna('Unknown', inplace=True)

df['DrugName'].fillna('Unknown', inplace=True)

df['SubCat'].fillna('Unknown', inplace=True)

df['SubCat1'].fillna('Unknown', inplace=True)
```

```python
categorical_cols = ['Formulation', 'DrugName', 'SubCat', 'SubCat1']

df = pd.get_dummies(df, columns=categorical_cols)
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
numerical_cols = ['Quantity', 'ReturnQuantity', 'Final_Cost', 'Final_Sales']

for col in numerical_cols:

    plt.hist(df[col], bins=50)

    plt.title(f'Histogram of {col}')

    plt.show()
```

```python
for col in numerical_cols:

    plt.boxplot(df[col])

    plt.title(f'Boxplot of {col}')
```

```python
corr_matrix = df.corr()

print(corr_matrix)
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix')

plt.show()
```
